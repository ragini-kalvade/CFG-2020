```python
import matplotlib.pyplot as plt
import random
```

```python
#Data will consist of assignement_id, student_id, marks obtained, submission status
import numpy as np
dummy_dataset = []
for i in range(0,100):
    if(i < 60):
        dummy_dataset.append([1,i,random.randint(1,11),'YES'])
    else:
        dummy_dataset.append([2,i,random.randint(1,11),'NO'])
```

```python
assignment_id = 1 #FROM DATABASE
```

```python
visual_array = []
for val in dummy_dataset:
    if(val[0] == 1 and val[3] == "YES"):
        visual_array.append(val[2])
```

```python
fig, ax = plt.subplots(1,2,figsize=(15,5))
dim=np.arange(1,11,1)
plt.setp(ax, xticks=dim)
plt.xlim(1, 11)
ax[0].hist(visual_array, bins=10, alpha = 0.5, color = 'r',edgecolor='b',linewidth=2)
ax[0].set_xlabel("MARKS")
ax[0].set_ylabel("NUMBER OF STUDENTS")
ax[0].set_title("WITH INTERNET")
ax[1].hist(visual_array[10:90], bins=10, alpha = 0.5, color = 'g',edgecolor='b',linewidth=2)
ax[1].set_xlabel("MARKS")
ax[1].set_ylabel("NUMBER OF STUDENTS")
ax[1].set_title("WITHOUT INTERNET/SMS")
plt.show()
```
